# JOINS

```python

customers = pd.DataFrame({
    "CustomerID": [1, 2, 3, 4],
    "CustomerName": ["Harish", "Sudha", "Ravi", "Meena"],
    "City": ["Chennai", "Bangalore", "Hyderabad", "Pune"]
})


orders = pd.DataFrame({
    "OrderID": [101, 102, 103, 104, 105],
    "CustomerID": [1, 2, 2, 5, 3],
    "Amount": [500, 700, 300, 400, 900]
})

customers
```

```python
orders
```

### INNER JOIN - Only consider matching records/rows in both the tables

```python
inner_join = pd.merge(customers ,orders , on= "CustomerID" , how = "inner" )


inner_join
```

### LEFT JOIN - Consider everythign from left , but only matching ones in right


```python

left_join = pd.merge(customers ,orders , on= "CustomerID" , how = "left" )


left_join
```

### RIGHT JOIN - Consider everythign from right table , but only matching ones in left


```python

right_join = pd.merge(customers ,orders , on= "CustomerID" , how = "right" )


right_join
```

### OUTER JOIN - Consider everythign from both left and right


```python

outer_join = pd.merge(customers ,orders , on= "CustomerID" , how = "outer" )


outer_join.loc[ outer_join["Amount"]>300 , ["City"]]

# Attributes
# Methods

outer_join
```
